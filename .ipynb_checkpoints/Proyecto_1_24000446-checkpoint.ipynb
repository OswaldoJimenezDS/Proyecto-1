```python
# Función para obtener escala de variables
def get_df_variable_scale_type(dataset):
    continuas = []
    discretas = []
    categoricas = []

    for col in dataset.columns:
        tipoDato = dataset[col].dtype
        if ((tipoDato == 'int64') or (tipoDato == 'float64')):
            if(len(dataset[col].unique()) > 30):
                continuas.append(col)
            else:
                discretas.append(col)
        else:
            categoricas.append(col)
    return continuas, discretas, categoricas
```

### 6) Creación de función para entrenar el modelo de regresión lineal de una variable 

```python
class regresion_lineal():
    def __init__(self, x, y, epochs, imprimir_error_cada, learning_rate):
        
        #Asignación de variables
        #Betas
        self.b0 = 0
        self.b1 = 0
        
        self.imprimir_error_cada = imprimir_error_cada
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.x_ini = x
        self.y_ini = y
        self.VUnos = np.ones_like(self.x_ini).reshape(-1,1)
        self.VXUnos = np.hstack([np.reshape(self.x_ini, (-1,1)),self.VUnos])
        self.VYe = np.reshape(self.y_ini, (-1,1))
        self.VError = []
        self.Model = {}
        self.betas = np.reshape([[self.b1],[self.b0]],(-1,1))

        for ep in range(epochs):
            self.yhat = np.dot(self.VXUnos, self.betas)
            self.Error = 1/(2*self.VYe.shape[0]) * sum((self.VYe - self.yhat)**2)
            self.VError.append(self.Error[0])
            if ep%round(imprimir_error_cada) == 0:
                print(f'\nIteraciones: {ep} Error: {self.Error[0]}')
            self.gradiente = np.reshape(learning_rate*(1/self.VYe.shape[0]*sum((self.yhat - self.VYe) * self.VXUnos)),(-1, 1))
            self.betas = self.betas - self.gradiente
            self.Model[ep] = [self.betas, self.yhat, self.VYe, self.x_ini]
        
    def model_reg(self):
        return self.Model, np.array(self.VError)

    #@staticmethod
    def errores(error):
        fig_a = plt.figure(figsize = (5, 5))
        ax = fig_a.add_axes([0.02, 0.02, 0.6, 0.6])
        ax.plot(error, color="#006400")
        ax.set_title("Errores del modelo")
        ax.set_xlabel("Numero de iteraciones")
        ax.set_ylabel("Error")
        plt.show()

    def evolucion_modelo(Modelo, n):
        for i in list(Modelo.keys()):
            if i%round(n) == 0:
                fig_b = plt.figure(figsize = (5, 5))
                ax2 = fig_b.add_axes([0.02, 0.02, 0.6, 0.6])
                yhat = Modelo[i][1]
                y = Modelo[i][2]
                x = Modelo[i][3]
                ax2.scatter(x, y, edgecolor='k', color="#8FBC8F")
                ax2.plot(x, yhat, linestyle='solid', color='r', label='Iteración:  %d' % i)
                ax2.set_title('Iteración: '+str(i))
                ax2.legend()
                ax2.set_xlabel("X")
                ax2.set_ylabel("Y")
                plt.show()

    def prediction(self, M):
        self.M = M
        self.VUnos = np.ones_like(self.M).reshape(-1,1)
        self.VXUnos = np.hstack([np.reshape(self.M, (-1,1)),self.VUnos])
        for i in range(self.epochs):
            self.Y_pred = np.dot(self.VXUnos, self.betas)
        return self.Y_pred

    def compare_prediction(MProp, Msklearn, Predict, Model):
        MPropR = MProp.prediction(Predict.reshape(-1, 1))
        MsklearnR = Msklearn.predict(Predict.reshape(-1, 1))
        Promedio = (MPropR + MsklearnR)/2
        fig_c = plt.figure(figsize = (5,5))
        ax3 = fig_c.add_axes([0.02, 0.02, 0.6, 0.6])
        for i in list(Model.columns.values):
            y = Model[i]
            ax3.scatter(Predict, y, edgecolor='k', color="yellow")
            ax3.plot(Predict, MPropR, color='red', label='Reg.Propia: ')
            ax3.plot(Predict, MsklearnR, color='green', label='Msklearn: ')
            ax3.plot(Predict, Promedio, color='blue', label='Promedio: ')
            ax3.set_title("Comparación")
            ax3.legend()
            ax3.set_xlabel("X")
            ax3.set_xlabel("Y")
            plt.show()
        return MPropR, MsklearnR, Promedio

    def compare_errors(Models1, Models2, Prueba1, Prueba2, Sprueba):
        labels = ['OverallQual', '1stFlrSF']
        Error_Reg_Prop = metrics.mean_squared_error(Models1[0], Prueba1)
        Error_Msklearn = metrics.mean_squared_error(Models1[1], Prueba1)
        Error_Promedio = metrics.mean_squared_error(Models1[2], Prueba1)

        Error2_Reg_Prop = metrics.mean_squared_error(Models2[0], Prueba2)
        Error2_Msklearn = metrics.mean_squared_error(Models2[1], Prueba2)
        Error2_Promedio = metrics.mean_squared_error(Models2[2], Prueba2)

        Errores_Reg_Prop = [Error_Reg_Prop, Error2_Reg_Prop]
        Errores_Msklearn = [Error_Msklearn, Error2_Msklearn]
        Errores_Promedio = [Error_Promedio, Error2_Promedio]

        T = np.arange(len(labels))
        fig_d = plt.figure(figsize = (9, 7))
        ax = fig_d.add_axes([0.01, 0.01, 0.6, 0.6])
        ax.set_xticks(T + 0.90/3)
        ax.set_xticklabels(labels)
        ax.bar(T, Errores_Reg_Prop, width=0.3, edgecolor='k', color="#f7dc6f", label="Reg.Prop")
        ax.bar(T+0.30, Errores_Msklearn, width=0.3, edgecolor='k', color="#82e0aa", label="Msklearn")
        ax.bar(T+0.60, Errores_Promedio, width=0.3, edgecolor='k', color="#a569bd", label="Promedio")
        ax.legend()
        ax.set_title("Comparación")
        plt.show()   
```

### Iteraciones/Evolución - OverallQual

```python
SaleP = np.reshape(np.array(set_de_entrenamiento['SalePrice']), (-1, 1))
OverallQ = np.reshape(np.array(set_de_entrenamiento['OverallQual']), (-1, 1))
OnestFlrSF = np.reshape(np.array(set_de_entrenamiento['1stFlrSF']), (-1, 1))
epochs = 5001
Imprimir_cada = 1000
L_rate = 0.01
RegModelo = regresion_lineal(OverallQ, SaleP, epochs, Imprimir_cada, L_rate)
Rmodelos, Rerrores = RegModelo.model_reg()
regresion_lineal.errores(Rerrores)
regresion_lineal.evolucion_modelo(Rmodelos, Imprimir_cada)
```

### Iteraciones/Evolución - 1stFlrSF

```python
epochs = 51
Imprimir_cada = 10
L_rate = 0.0000001
RegModelo2 = regresion_lineal(OnestFlrSF, SaleP, epochs, Imprimir_cada, L_rate)
Rmodelos1, Rerrores1 = RegModelo2.model_reg()
regresion_lineal.errores(Rerrores1)
regresion_lineal.evolucion_modelo(Rmodelos1, Imprimir_cada)
```
